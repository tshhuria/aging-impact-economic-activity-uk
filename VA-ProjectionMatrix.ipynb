```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

economic_status_2021_path = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/EconomicStatus-2021.csv'
household_composition_2021_path = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/HouseholdComposition-2021.csv'
economic_activity_2011_path = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/EconomicActivity-2011.csv'
household_composition_2011_path = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/HouseholdComposition-2011.csv'

economic_status_2021 = pd.read_csv(economic_status_2021_path)
household_composition_2021 = pd.read_csv(household_composition_2021_path)
economic_activity_2011 = pd.read_csv(economic_activity_2011_path)
household_composition_2011 = pd.read_csv(household_composition_2011_path)

data_2021 = pd.merge(economic_status_2021, household_composition_2021, on='geography code')
data_2011 = pd.merge(economic_activity_2011, household_composition_2011, on='geography code')

# Function to perform PCA projection
def perform_pca(data, n_components=2):
    pca = PCA(n_components=n_components)
    components = pca.fit_transform(data.select_dtypes(include=[float, int])) # Select numerical columns
    return pd.DataFrame(components, columns=[f'PCA{dim+1}' for dim in range(n_components)])

# Function to perform t-SNE projection
def perform_tsne(data, n_components=2):
    tsne = TSNE(n_components=n_components)
    components = tsne.fit_transform(data.select_dtypes(include=[float, int])) # Select numerical columns
    return pd.DataFrame(components, columns=[f'tSNE{dim+1}' for dim in range(n_components)])

# Perform projections for 2011 data
pca_2011 = perform_pca(data_2011)
tsne_2011 = perform_tsne(data_2011)

# Perform projections for 2021 data
pca_2021 = perform_pca(data_2021)
tsne_2021 = perform_tsne(data_2021)

# Save the projections to CSV files
pca_2011.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/PCA_2011.csv', index=False)
tsne_2011.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/tSNE_2011.csv', index=False)
pca_2021.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/PCA_2021.csv', index=False)
tsne_2021.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021//tSNE_2021.csv', index=False)
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Paths to the CSV files
tsne_path_2011 = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/tSNE_2011.csv'
pca_path_2011 = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/PCA_2011.csv'
tsne_path_2021 = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/tSNE_2021.csv'
pca_path_2021 = '/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/PCA_2021.csv'

# Load the CSV files
tsne_2011 = pd.read_csv(tsne_path_2011)
pca_2011 = pd.read_csv(pca_path_2011)
tsne_2021 = pd.read_csv(tsne_path_2021)
pca_2021 = pd.read_csv(pca_path_2021)

# Apply K-means clustering
kmeans_tsne_2011 = KMeans(n_clusters=5).fit(tsne_2011[['tSNE1', 'tSNE2']])
kmeans_pca_2011 = KMeans(n_clusters=5).fit(pca_2011[['PCA1', 'PCA2']])
kmeans_tsne_2021 = KMeans(n_clusters=5).fit(tsne_2021[['tSNE1', 'tSNE2']])
kmeans_pca_2021 = KMeans(n_clusters=5).fit(pca_2021[['PCA1', 'PCA2']])
```

```python
tsne_2011['Cluster'] = kmeans_tsne_2011.labels_
pca_2011['Cluster'] = kmeans_pca_2011.labels_
tsne_2021['Cluster'] = kmeans_tsne_2021.labels_
pca_2021['Cluster'] = kmeans_pca_2021.labels_

# Save the updated CSV files with new names to avoid overwriting issues
tsne_2011.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/tSNE_2011_clustered.csv', index=False)
pca_2011.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2011/PCA_2011_clustered.csv', index=False)
tsne_2021.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/tSNE_2021_clustered.csv', index=False)
pca_2021.to_csv('/Users/tanmaysagarhuria/Desktop/VisualAnalyticsCourseWork/2021/PCA_2021_clustered.csv', index=False)
```
